```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('cip_dataset_v2.csv')
df['date']=pd.to_datetime(df['date'])
df.set_index('date',inplace=True)
```

```python
AUD=df[(df['currency']=='AUD') & (df['tenor']=='3m') & (df.index.year>=2000)]
EUR=df[(df['currency']=='EUR') & (df['tenor']=='3m')& (df.index.year>=2000)]
JPY=df[(df['currency']=='JPY') & (df['tenor']=='3m')& (df.index.year>=2000)]
plt.figure(figsize=(12,6))
plt.axhline(0)
plt.axvline(pd.datetime(2008,9,15), color='k', linestyle='dashed')
plt.axvline(pd.datetime(2011,10,26), color='k',linestyle='dashed')

plt.plot(-AUD['cip_govt'],label='AUD',color='r')
plt.plot(-EUR['cip_govt'],label='EUR',color='b')
plt.plot(-JPY['cip_govt'],label='JPY',color='g')
plt.ylim(-200,200)
plt.ylabel('basis points')
plt.title('3-month Basis against USD', fontsize=20)
plt.legend(loc='upper right', fontsize=20)
```

```python
AUD=df[(df['currency']=='AUD') & (df['tenor']=='3y') & (df.index.year>=2000)]
EUR=df[(df['currency']=='EUR') & (df['tenor']=='3y')& (df.index.year>=2000)]
JPY=df[(df['currency']=='JPY') & (df['tenor']=='3y')& (df.index.year>=2000)]
plt.figure(figsize=(12,6))
plt.axhline(0)
plt.axvline(pd.datetime(2008,9,15), color='k', linestyle='dashed')
plt.axvline(pd.datetime(2011,10,26), color='k',linestyle='dashed')

plt.plot(-AUD['cip_govt'],label='AUD', color='r')
plt.plot(-EUR['cip_govt'],label='EUR', color='b')
plt.plot(-JPY['cip_govt'],label='JPY', color='g')
plt.ylim(-200,200)
plt.ylabel('basis points')
plt.title('3-year Basis against USD', fontsize=20)
plt.legend(loc='upper right', fontsize=20)
```

```python
ted_df=pd.read_csv('ted-spread.csv')
ted_df['date']=pd.to_datetime(ted_df['date'], dayfirst=True)
ted_df.set_index('date', inplace=True)
```

```python
ted_df=ted_df[(ted_df.index.year>=2000)& (ted_df.index.year<=2019)]
plt.figure(figsize=(12,6))

plt.plot(ted_df[' value'],label='3m Ted-spread', color='r')
plt.axvline(pd.datetime(2008,9,15), color='k', linestyle='dashed')
plt.axvline(pd.datetime(2011,10,26), color='k',linestyle='dashed')
plt.title('3m Ted-spread during crisis', fontsize=20)
plt.ylabel('% points')
plt.legend()
```
